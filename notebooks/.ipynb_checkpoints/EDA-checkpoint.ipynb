```python
import numpy as np 
import pandas as pd

df = pd.read_csv("/Users/tramy/Documents/Github2/bigco/data/raw/employee_reviews.csv")

df.head()
```

1. Count the number of reviews given to each company.

```python
from matplotlib import pyplot as plt
import seaborn as sns

#1.Count the number of reviews given to each company using seaborn lib

sns.countplot(x=df['company'], data=df)
plt.yticks(np.arange(0, 27000, 2000))
```

2. Plot the distribution of overall ranking across companies

```python
#2. Plot the distribution of overall ranking across companies
# Need to set "as_index = False", otherwise cannot plot.
# Set up plot
figsize=(20, 16)
ticksize = 10
titlesize = ticksize + 3
labelsize = ticksize + 3

sns.set(context='notebook', style='whitegrid', palette='deep', font='sans-serif', font_scale=1, color_codes=True, rc=None)
sns.set_style("ticks", {"xtick.major.size": ticksize, "ytick.major.size": ticksize})


df2 = df.groupby('company', as_index = False)['overall-ratings'].mean()
print(df2)
colx = 'company'
coly = 'overall-ratings'
sns.barplot(x= colx, y = coly, data = df2) 
plt.yticks(np.arange(0, 5, 0.5))
plt.title("Number of Reviews across Companies")
```

3. Plot distribution of all other types of rankings

```python
#3. Plot distribution of all other types of rankings

ranking = ['overall-ratings','work-balance-stars','culture-values-stars','carrer-opportunities-stars','comp-benefit-stars','senior-mangemnet-stars']

# Make sure all data is numerical
df[ranking] = df[ranking].apply(pd.to_numeric, errors='coerce')

nrows= 3 #specify for later use = number and display of subplots
ncols= 2
figsize = (20,16)

fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=figsize) #adjust the size of the figure
plt.subplots_adjust(hspace=.6, wspace=.3)

for i in range(0,len(ranking)):
    plt.subplot(nrows,ncols,i+1)
    df3 = df.groupby('company', as_index=False)[ranking[i]].mean() #calculate the mean
    xcol = 'company'
    ycol = ranking[i]
    sns.barplot(x= xcol, y = ycol, data = df3) 
    plt.yticks(np.arange(0, 5, 0.5))
    plt.title(ranking[i])
    plt.plot()
    
```

As seen above, Facebook seems to have the best ratings compared to others in all aspects. However, as previously shown, Facebook also has the least number of reviews. The data result could be bias in that sense.

4. Distribution of job titles and employment status of all reviewers.

```python
# Define the function that gets the employment status of the reviewers
def get_stt(x):
    if "(" in x:
        x = x.split("- ")[0]
    if "- " in x:
        x = x.split("- ")[0]
    return x

get_stt("Current Employee - Anonymous Employee")
 
df["job-title"] = df["job-title"].apply(lambda x: get_stt(x))

sns.countplot(x = df['job-title'], data=df)
```

5. Distribution of ranking of one company in different locations

```python
#Get the location
def get_state(x):
    if "(" in x:
        x = x.split("(")[1]
        x = x.split(")")[0]
    elif ", " in x:
        x = x.split(", ")[-1]
    return x

df["state"] = df["location"].apply(lambda x: get_state(x))
```

```python
df_comp = df.groupby("company", as_index=False)
df_amz = df_comp.get_group('amazon') #take out group amazon

df_a1 = df_amz.groupby("state", as_index=False)["overall-ratings"].mean()

#Plot

figsize=(20, 8)

fig, axs = plt.subplots(nrows=1, ncols=1, figsize=figsize)
plt.subplots_adjust(hspace=1, wspace=.3)
sns.barplot(x = "state", y = "overall-ratings", data= df_a1.head(50))
plt.xticks(rotation=90)
```
