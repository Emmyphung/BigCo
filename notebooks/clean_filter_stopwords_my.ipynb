Clean, tokenize and filter text data from stop words


```python
df['summary'] = df['summary'].astype(str)
df['summary']= df['summary'].str.lower().str.split()

df['cons'] = df['cons'].astype(str)
df['cons']= df['cons'].str.lower().str.split()

df['pros'] = df['pros'].astype(str)
df['pros']= df['pros'].str.lower().str.split()

df['advice-to-mgmt']= df['advice-to-mgmt'].astype(str)
df['advice-to-mgmt']= df['advice-to-mgmt'].str.lower().str.split()
```

```python
df['pros']
```

```python
import re
#pattern = re.compile(r"[^\w]+[\d]") #set a pattern to filter data with

def non_word_pull_regex(str):
    pull = re.findall(r'\d', str)
    pull2 = re.findall(r'\W', str)
    return pull +pull2

def determine_if_string_has_non_word_chars(item):
    boolean_result = len(non_word_pull_regex(item)) ==0
    return boolean_result

def apply_function_to_clean_data(data,func):
    result = [[item for item in sublist if func(item)] for sublist in data]
    return result
```

```python
df['summary_clean'] = apply_function_to_clean_data(df['summary'],determine_if_string_has_non_word_chars)
df['pros_clean'] = apply_function_to_clean_data(df['pros'],determine_if_string_has_non_word_chars)
df['cons_clean'] = apply_function_to_clean_data(df['cons'],determine_if_string_has_non_word_chars)
df['advice_clean'] = apply_function_to_clean_data(df['advice-to-mgmt'],determine_if_string_has_non_word_chars)
```

```python
df['cons_clean']
```

```python
#Filter stop words

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
stop_words.add('i\'ve')
```

```python
def filter_stop_words(data):
    result = [[word for word in row if word not in stop_words] for row in data]
    return result
```

```python
df['summary_filtered'] = filter_stop_words(df['summary_clean'])
df['pros_filtered']= filter_stop_words(df['pros_clean'])
df['cons_filtered']= filter_stop_words(df['cons_clean'])
df['advice_filtered']= filter_stop_words(df['advice_clean'])
```
