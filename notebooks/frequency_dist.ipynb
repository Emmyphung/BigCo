Frequency Distribution

1. See which keywords are most frequently mentioned per company.

```python
import pandas as pd
import numpy as np
import nltk
```

```python
df = pd.read_csv("/Users/tramy/Documents/GitHub/bigco/data/interim/employee_reviews_tokenized_filtered.csv")
```

```python
np.unique(df['company'])
```

```python
df['filtered_summary_word'].head(10)
```

```python
# Calculate frequency distribution
from nltk.probability import FreqDist
summary_review = df['filtered_summary_word']
fdist_summary = nltk.FreqDist(summary_review)

# Output top 50 words

for word, frequency in fdist_summary.most_common(50):
    print(u'{};{}'.format(word, frequency))
```

```python
import matplotlib.pyplot as plt
fdist_summary.plot(30,cumulative=False)
plt.show()
```
